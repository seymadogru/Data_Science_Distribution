```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

my_normal_dagilimli = np.random.normal(70, 200, 100)
my_normal_dagilimli.sort()
plt.figure(figsize=(6,3), dpi = 100)
plt.plot(my_normal_dagilimli,":",color="darkred") 
plt.ylabel('Normal Dağılımlı', fontsize = 15, color= 'darkgreen') 
plt.show()
```
